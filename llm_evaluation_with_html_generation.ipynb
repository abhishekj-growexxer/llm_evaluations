# Merge all models predictions in HTML format

```python
import re

def prepare_for_html(text):
    # Replace newline characters (\n) with <br> tags
    html_text = re.sub(r'\n', '<br>', text)
    
    # Add more conversions for other tags here
    html_text = re.sub(r'\*\*(.*?)\*\*', r'<strong>\1</strong>', html_text)  # Convert **text** to <strong>text</strong>
    html_text = re.sub(r'__(.*?)__', r'<em>\1</em>', html_text)  # Convert __text__ to <em>text</em>
    # Add more conversions as needed for other tags
    
    # Check for repetition and pick the first occurrence of the repeated tag
    seen_tags = set()
    def pick_first(match):
        tag = match.group(0)
        if tag in seen_tags:
            return ""
        seen_tags.add(tag)
        return tag

    html_text = re.sub(r'<[^>]+>', pick_first, html_text)
    
    return html_text
```

```python
html = """<tr>
            <td>{}</td> 
            <td>{}</td>
            <td>{}</td>
            <td class="model-col">{}</td>
            <td class="model-col">{}</td>
            <td class="model-col">{}</td>
            <td class="model-col">{}</td>
            <td class="model-col">{}</td>
            <td class="model-col">{}</td>
            <td class="model-col">{}</td>
            <td class="model-col">{}</td>
            <td>{}</td>
        </tr>"""

final_html = ''

for i in range(len(questions)):
    
    question = prepare_for_html(questions[i])
    
    correct_answer = prepare_for_html(correct_answers[i])
    
    question_type = que_type[i+1]
    
    flan_t5_ans = prepare_for_html(flan_df['flan-t5'][i])
    
    openllama_ans = prepare_for_html(openllama_df['openllama'][i])
    
    alpaca_ans = prepare_for_html(alpaca_df['alpaca'][i])
    
    llama_ans = prepare_for_html(llama_df['llama'][i])
    
    vicuna_ans = prepare_for_html(vicuna_df['vicuna'][i])
    
    openchat_ans = prepare_for_html(openchat_df['openchat'][i])
    
    llama2_ans = prepare_for_html(llama2_df['llama2'][i])

    gpt_ans = prepare_for_html(gpt_df['gpt'][i])

    final_html += ' ' + html.format(i+1, question, question_type ,flan_t5_ans, openllama_ans  ,alpaca_ans, llama_ans, vicuna_ans ,openchat_ans, llama2_ans, gpt_ans, correct_answer)
    

# Last cleaning process
final_html = re.sub(r'\n', ' ', final_html)
print(final_html)
```
